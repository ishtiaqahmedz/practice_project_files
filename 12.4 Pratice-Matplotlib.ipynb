# A note on the Object-Oriented API vs Pyplot
Matplotlib has two interfaces. The first is an object-oriented (OO) interface. In this case, we utilize an instance of axes.Axes in order to render visualizations on an instance of figure.Figure.

The second is based on MATLAB and uses a state-based interface. This is encapsulated in the pyplot module.'''

The Figure is the final image that may contain 1 or more Axes.
The Axes represent an individual plot (don't confuse this with the word "axis", which refers to the x/y axis of a plot).

```python
hd=pd.read_csv("heart-disease.csv")
hd.head(10)
```

```python
hd["age"].hist(bins=20);
```

```python
hd.plot.hist(figsize=(10,30),subplots=True,bins=30);
```

```python
over_50=hd[
        hd["age"]>50
        ]

over_50
```

```python
#pyplot methods are used above and we are doing it again:

over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(4,3),
            );

```

```python

#OO method:

fig,ax=plt.subplots(figsize=(15,4))

over_50.plot(kind='scatter',
            x="age",
            y="chol",
            c='target',
            ax=ax);
ax.set_xlim(40,100) #increase to x axis limit from 50-90 to 40-100
```

```python
#building OO plot from scratch 
fig,ax=plt.subplots(figsize=(10,4))

scatter=ax.scatter(over_50["age"],
                   over_50["chol"],
                  c=over_50["target"])

ax.set(title="Age and Cholestrol Level",xlabel="AGE",ylabel="CHOLESTROL")
ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over_50["chol"].mean(),linestyle='---',color='r')
```

```python
fix,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                          sharex=True)


scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

ax0.axhline(y=over_50["chol"].mean(),linestyle="--",label="Average")

ax0.legend(*scatter.legend_elements(),title="Target")

ax0.set(title="Age and Cholestrol Levels",ylabel="Cholestrol")



scatter1=ax1.scatter(x=over_50["age"],
           y=over_50["thalach"],
           c=over_50["target"])

ax1.set(title="Age and Max Heart Rate",xlabel="Age",ylabel="Max Heart Rate")

ax1.legend(*scatter1.legend_elements(),title="Target")
ax1.axhline(over_50["thalach"].mean(),linestyle="--",color='r');

fig.suptitle('Heart Diesease Analysis', fontsize=15,fontweight='bold');

#fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
